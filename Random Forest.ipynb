```python
# Encoding - convert  categorical columns data to Numerical data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
# Create a list with all object data type column so you canc onvert them to numerical

col_list =[ ]

for i in churn_data.columns:
  if ((churn_data[i].dtype =="object") & (i!="Churn")):
    col_list.append(i)

col_list
```

```python
for i in col_list:
  churn_data[i]=le.fit_transform(churn_data[i])
```

```python
# Dividing the data into Independent variables and dependent varibale ( Target)

X=churn_data.iloc[: , :-1]
y= churn_data["Churn"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test =  train_test_split( X,y , test_size = 0.2 , random_state=0 , stratify=y)
```

```python
from random import Random

from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model1 = RandomForestClassifier(n_estimators=100)  # Number of decision trees that you want to use = 100
```

```python
# Train the Random forest model
RF_model1.fit(X_train , y_train)
```

```python
# Test the Random forest model
ans = RF_model1.predict(X_test)
```

```python
ans
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(ans , y_test)
```

```python
cm = confusion_matrix(ans ,y_test)
cm
```

```python
#  Use Different values of n_estimator to find which optimal no of DT is going to give us highest accuracy in Random Forest

n=[50,100,200,300,400,500,600,700,800,900,1000]

accuracy = [ ]   # In this list i am going to store all the accuracy

for i in n:
  model = RandomForestClassifier(n_estimators=i)
  model.fit(X_train , y_train)                      # Training the model
  ans =model.predict(X_test)                        # Testing the model
  a=accuracy_score(y_test,ans)*100
  accuracy.append(a)
```
